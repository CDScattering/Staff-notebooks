```python
def get_ring(ar_sz, r):
    ring_mask = np.abs(tb.genAp(ar_sz, r+3)) * np.abs(tb.genStop(ar_sz, r))
    return ring_mask
```

```python
%matplotlib qt5
with h5py.File(filepath, 'r') as f:
    try:
        probe = f['entry_1']['process_1']['output_1']['probe']
        probe = np.squeeze(np.array(probe))
    except:
        probe = np.zeros([1, 1])

    try:
        obj = f['entry_1']['process_1']['output_1']['object']
        obj = np.squeeze(np.array(obj))
    except:
        obj = np.zeros([1, 1])

    try:
        scan_pos = f['entry_1']['process_1']['output_1']['scan_positions']
        scan_pos = np.squeeze(np.array(scan_pos))
    except:
        scan_pos = np.zeros([2, 1])

    try:
        scan_tilt = f['entry_1']['process_1']['output_1']['scan_tilts']
        scan_tilt = np.squeeze(np.array(scan_tilt))
    except:
        scan_tilt = np.zeros([2, 1])

    try:
        wav = np.array(f['entry_1']['experiment_1']['source']['wavelength'])[0]
    except:
        wav = 1
        
    try:
        dx = np.array(f['entry_1']['process_1']['common_1']['dx'])[0][0]
    except:
        dx =1
        
    d_theta = wav/(obj.shape[-1] * dx)
    
    d_spacing = 2.35e-10
    bragg = (wav/(1*d_spacing))
    
    print("wav", wav)
    print("dx", dx)
    print("dt", d_theta)
    print("bragg angle", bragg)
    ring_mask = get_ring(obj.shape,bragg/d_theta)
        
    probe_ft = np.fft.fftshift(np.fft.fft2(probe))
    obj_ft = np.fft.fftshift(np.fft.fft2(obj))

    plt.figure('PtyOUT', figsize=[15, 15])
    plt.gray()

    plt.subplot(2, 3, 1)
    plt.imshow(np.abs(tb.tile(probe)))
    plt.title('probe modulus')

    plt.subplot(2, 3, 4)
    plt.imshow(np.log(np.abs(tb.tile(probe_ft))))
    plt.title('probe ft [log]')

    plt.subplot(2, 3, 2)
    plt.imshow(np.angle(tb.tile(obj)))
    plt.title('object phase')

    plt.subplot(2, 3, 5)
    obj_ft[ring_mask>=1]=np.amax(obj_ft)
    plt.imshow(np.log(np.abs(tb.tile(obj_ft))))
    plt.title('object ft [log]')

    plt.subplot(2, 3, 3)
    plt.imshow(np.abs(tb.tile(obj)))
    plt.title('object modulus')
    
    plt.subplot(2, 3, 6)
    plt.imshow(np.abs(ring_mask))
    plt.title('ring')
```
