```python
%matplotlib notebook
from ase import Atoms
from ase.build import mx2
from ase.visualize.plot import plot_atoms
import matplotlib.pyplot as plt
from ase.visualize import view, ngl#
import nglview
```

```python
#build structure
MoS2 = mx2(formula='MoS2', kind='2H', a=3.18, thickness=3.19, size=(3, 3, 1), vacuum=(2))
```

```python
#visualising with Matplotlib
fig, axarr = plt.subplots(1, 2, figsize=(8, 5))
fig2, axarr2 = plt.subplots(1, 2, figsize=(8, 5))
plot_atoms(MoS2, axarr[0], radii=0.3, rotation=('0x,0y,0z'))
plot_atoms(MoS2, axarr[1], scale=0.7, offset=(3, 4), radii=0.3, rotation=('0x,0y,0z'))
plot_atoms(MoS2, axarr2[0], radii=0.3, rotation=('45x,45y,0z'))
plot_atoms(MoS2, axarr2[1], radii=0.3, rotation=('0x,0y,0z'))
axarr[0].set_title("No rotation")
axarr[1].set_title("Offset")
axarr[1].set_xlabel("X-axis, [$\mathrm{\AA}$]")
axarr[1].set_ylabel("Y-axis, [$\mathrm{\AA}$]")
axarr2[0].set_axis_off()
axarr2[0].set_title("45deg rotation")
#axarr[3].set_xlim(2, 6)
#axarr[3].set_ylim(2, 6)
```
